# Analisa Status Pekerjaan Penduduk Miskin

Data Status Pekerjaan Penduduk Miskin bersumber dari http://data.go.id. 
Pada data ini terdapat kolom kode_provinsi, nama_provinsi, kelompok_usia,status_kerja,dan jumlah_individu

```python
import pandas as pn
data_pekerjaan = pn.read_csv('processed-jumlah-status-bekerja-30persen-termiskin-indonesia-perjuli2012.csv', delimiter = ',', encoding = "ISO-8859-1")
```

```python
print("Jumlah Data : ", data_pekerjaan.count())
data_pekerjaan
```

```python
s = data_pekerjaan.jumlah_individu.sum()
me = data_pekerjaan.jumlah_individu.mean()
md = data_pekerjaan.jumlah_individu.median()



print("Jumlah total penduduk miskin :",s,"orang")
print("Jumlah terbesar penduduk miskin :",ma,"orang")
print("Jumlah terkecil penduduk miskin :",mi,"orang")
print("Mean Data : ",me)
print("Median Data :",md)
```

```python
jml_prov=data_pekerjaan.groupby('nama_provinsi').sum()['jumlah_individu']
print("Jumlah Data Penduduk Miskin berdasarkan Provinsi")
print("=====================================================")
print(jml_prov.sort_values())
print("\n")
print("Rata-rata :",jml_prov.mean())
print("Jumlah :",jml_prov.sum())

ax = jml_prov
ax = ax.plot(kind='bar', title ="Jumlah Data Penduduk Miskin berdasarkan Provinsi", legend=True, fontsize=12)
ax.set_xlabel("Provinsi", fontsize=12)
ax.set_ylabel("Jumlah", fontsize=12)

```

```python
usia=data_pekerjaan.groupby('kelompok_usia').sum()['jumlah_individu']
print("Jumlah Data Penduduk Miskin yang bekerja berdasarkan Kelompok Usia")
print("==========================================================")
print(usia)

ax = usia
ax = ax.plot(kind='bar', title ="Jumlah Data Penduduk Miskin berdasarkan Kelompok Usia", legend=True, fontsize=12)
ax.set_xlabel("Usia", fontsize=12)
ax.set_ylabel("Jumlah", fontsize=12)
```

```python
status=data_pekerjaan.groupby('status_kerja').sum()['jumlah_individu']
print("Jumlah Data Penduduk Miskin yang bekerja berdasarkan Status Kerja")
print("==========================================================")
print(status)

ax = status
ax = ax.plot(kind='bar', title ="Jumlah Data Penduduk Miskin berdasarkan Status Kerja", legend=True, fontsize=12)
ax.set_xlabel("Status Kerja", fontsize=12)
ax.set_ylabel("Jumlah", fontsize=12)
```
