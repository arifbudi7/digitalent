# LATIHAN 2 - Tugas List Nilai

Diberikan sekumpulan Nilai Random antara 0 sampai dengan 100 sebanyak 250 nilai. Buatlah program untuk melakukan hal berikut:
1. Tampilkan Angka Ganjil dari ke 250 nilai tersebut
2. Tampilkan berapa banyak yang mendapatkan nilai lebih dari 70
3. Tampilkan standar deviasi dari sekumpulan nilai tersebut
4. Tampilkan nilai maksimum dan minimum
5. Tampilkan modus dari sekumpulan nilai tersebut

```python
import random
import statistics 

# GENERATE RANDOM
nilaiTugas = []
for i in range(250):
    nilaiTugas.append(random.randrange(1,100,1))
    
# TAMPILKAN:
# ganjil
print("Jawaban Soal 1 : Nilai Ganjil")
for ganjil in nilaiTugas:
    if ganjil % 2 != 0:
        print(ganjil,end=' ')
        
# nilai > 70
print("")
print("\nJawaban Soal 2 : Nilai > 70")
for nilai in nilaiTugas:
    if nilai > 70:
        print(nilai,end=' ')
        
        
# standar deviasi        
print("")
print("\nJawaban Soal 3 : Standar Deviasi")    
#nilai rata-rata
nilaiTotal = 0
for n in nilaiTugas:
    nilaiTotal = nilaiTotal + n
rata = nilaiTotal / len(nilaiTugas)
#standarDeviasi
selisih=0
for n in nilaiTugas:
    x=n-rata
    pangkat=x**2
    selisih=selisih+pangkat
standarDeviasi=(selisih/(len(nilaiTugas)-1))**(0.5)
print("Standard Deviasi=",standarDeviasi)

# maksimum
print("\nJawaban Soal 4 : Nilai Maksimum dan Minimum")  
max=0
for n in nilaiTugas:
    if n>max:
        max=n
print("Nilai Maksimum=",max)
# minimum 
for n in nilaiTugas:
    if n<max:
        max=n
print("Nilai Minimum =",max)


# modus
print("\nJawaban Soal 5 : Modus") 
parameterModus= 0
nilaiModus = []
angka = nilaiTugas[0]
banyakAngka = 1

for i in range(1, len(nilaiTugas)):
    if (nilaiTugas[i] == angka):
        banyakAngka += 1
    else:
        if (banyakAngka > parameterModus):
            parameterModus = banyakAngka
            nilaiModus = []
            nilaiModus.append(angka)
        elif(banyakAngka == parameterModus):
            nilaiModus.append(angka)
        angka = nilaiTugas[i]
        banyakAngka = 1
if (banyakAngka > parameterModus):
    parameterModus = banyakAngka
    nilaiModus = []
    nilaiModus.append(angka)
elif(banyakAngka == parameterModus):
    nilaiModus.append(angka)
print("Modus  =", nilaiModus)
```
